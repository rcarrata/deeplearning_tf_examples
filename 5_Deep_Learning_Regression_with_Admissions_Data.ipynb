```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import tensorflow as tf
from tensorflow	import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras import layers
from tensorflow.keras.layers import InputLayer



from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import Normalizer
from sklearn.compose import ColumnTransformer
from sklearn.metrics import r2_score
from tensorflow.keras.optimizers import Adam

df = dataset.copy()

features = df.iloc[:, 1:8]
labels = df.iloc[:,-1]

#print(features)
#print(labels)

# Because there are not categorical variables in this dataset
# the one-hot encoding is not required.
#features = pd.get_dummies(features)

features_train, features_test, labels_train, labels_test = train_test_split(features, labels, test_size=0.25, random_state=42)

sc = StandardScaler()

features_train_scaled = sc.fit_transform(features_train)
features_test_scaled = sc.fit_transform(features_test)

def design_model(features):
  #input = InputLayer(input_shape=(features.shape[1], ))
  num_features = features.shape[1]
  input = tf.keras.Input(shape=(num_features))
  model = Sequential(name="admissions_data")
  print(features.shape[1])
  model.add(input) # Input Layers
  model.add(layers.Dense(16, activation= 'relu')) # Hidden Layers
  model.add(layers.Dropout(0.1))
  model.add(layers.Dense(8, activation= 'relu')) # Hidden Layers
  model.add(layers.Dropout(0.2))
  model.add(layers.Dense(1)) # Output Layers
  
  opt = Adam(learning_rate=0.005)
  model.compile(loss='mse', metrics=['mae'], optimizer=opt)
  return model

model = design_model(features_train_scaled)
#print(model.summary())

num_epochs = 100
learning_rate=0.005
num_batch_size=8

es = EarlyStopping(monitor='val_loss', mode='min', verbose=1, patience=20)
history = model.fit(features_train_scaled, labels_train.to_numpy(), epochs=num_epochs, batch_size=num_batch_size, verbose=0, validation_split=0.25, callbacks=[es])

val_mse, val_mae = model.evaluate(features_test_scaled, labels_test.to_numpy(), verbose=0)

print("MAE: ", val_mae)

# evauate r-squared score
y_pred = model.predict(features_test_scaled)

print("R2 Score: ")
print(r2_score(labels_test,y_pred))

fig = plt.figure()
ax1 = fig.add_subplot(2, 1, 1)
ax1.plot(history.history['mae'])
ax1.plot(history.history['val_mae'])
ax1.set_title('model mae')
ax1.set_ylabel('MAE')
ax1.set_xlabel('epoch')
ax1.legend(['train', 'validation'], loc='upper left')

# Plot loss and val_loss over each epoch
ax2 = fig.add_subplot(2, 1, 2)
ax2.plot(history.history['loss'])
ax2.plot(history.history['val_loss'])
ax2.set_title('model loss')
ax2.set_ylabel('loss')
ax2.set_xlabel('epoch')
ax2.legend(['train', 'validation'], loc='upper left')

#plt.savefig('/content/drive/My Drive/Colab/Admissions/admissions.png')

#plot(history, '/content/drive/My Drive/Colab/Admissions/admissions.png')
```
